```python
import os
import shutil

source_path = r"C:\Users\KIIT\facial_expression"

os.makedirs("dataset/Stressed", exist_ok=True)
os.makedirs("dataset/Not_stressed", exist_ok=True)

stress_classes = ["angry", "fear", "sad","disgust"]
no_stress_classes = ["happy", "neutral","surprise"]

for folder in stress_classes:
    folder_path = os.path.join(source_path, folder)
    for img in os.listdir(folder_path):
        shutil.copy(
            os.path.join(folder_path, img),
            "dataset/Stressed"
        )

for folder in no_stress_classes:
    folder_path = os.path.join(source_path, folder)
    for img in os.listdir(folder_path):
        shutil.copy(
            os.path.join(folder_path, img),
            "dataset/Not_stressed"
        )

print("Images copied successfully")
```
